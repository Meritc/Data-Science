```python
#Corelation of features in the dataset?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Calculate the correlation matrix
correlation_matrix = data.corr()
# Create a heatmap for visualization
plt.figure(figsize=(13, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#Distribution of the target variable
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have already loaded the dataset into a Pandas DataFrame named 'data'
# If you haven't loaded the dataset yet, you can do so using pd.read_csv() or other appropriate methods.

# Assuming the target variable column name is 'target_variable'
target_variable = 'Y'

# Display data distribution in the target variable
plt.figure(figsize=(4, 4))
sns.countplot(x=target_variable, data=data)
plt.title('Data Distribution in Target Variable')
plt.xlabel('Target Variable')
plt.ylabel('Count')
plt.show()
```
